```python
import numpy as np
import matplotlib.pyplot as plt
#import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import pandas as pd
```

```python
plt.rcParams.update({'font.size': 10})
plt.rcParams['axes.linewidth'] = 1.5
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams.update({
    "text.usetex": True,
})
```

```python
def gaussian_beam(x, y, power, w_0=2.2e-3, wavelength=1550e-9):
    I_0 = 2*power/(np.pi*w_0**2)
    return I_0*np.exp(-2*(np.power(x,2)+np.power(y,2))/(w_0**2))

def wavefront_radius(z, z_0):
    return z*(1+np.power(z_0/z, 2))

def beam_radius(z, z_0, w_0):
    return w_0*np.sqrt(1+np.power(z/z_0, 2))

def phase_correct(z, z_0):
    return np.arctan(z/z_0)
```

```python

fig = plt.figure(figsize=(5.4*1.3*10, 10*4.2))
factor = 0.7
fig.set_size_inches(3*factor, 3*factor, forward=True)

x = np.linspace(-3.5e-3, 3.5e-3, 300)
X, Y = np.meshgrid(x, x.copy())
plt.pcolor(X/1e-3, Y/1e-3, gaussian_beam(X, Y, 300e-3), cmap=matplotlib.cm.viridis)
plt.xlabel(r'$x$ (mm)')
plt.ylabel(r'$y$ (mm)')
plt.xticks([-2, 0, 2])
plt.yticks([-2, 0, 2])
plt.savefig('../../plots/Chapter 2/gaussian_pcolor.pdf', dpi=300, bbox_inches = 'tight')

plt.show()

```

```python
colors = ['#2E86C1', '#85C1E9', '#1B4F72']

fig = plt.figure(figsize=(5.4*1.3*10, 10*4.2))
factor = 0.7
fig.set_size_inches(3.9*factor, 1*factor, forward=True)
x = np.linspace(-3e-3, 3e-3,1000)
plt.plot(np.linspace(-3e-3, 3e-3,1000), 
         gaussian_beam(x, 0, 300e-3, w_0=2.2e-3/2), 
         color=colors[2])
plt.savefig('../../plots/Chapter 2/gaussian_line.pdf', dpi=300, bbox_inches = 'tight')
plt.show()
```
