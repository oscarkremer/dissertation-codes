```python
colors = ['#2E86C1', '#85C1E9', '#1B4F72']

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)
factor = 0.7
fig.set_size_inches(6*factor, 5*factor, forward=True)
t = np.load("../../data/Chapter 3/Parametric Cooling/time_x.npy")
ph_001 = np.load("../../data/Chapter 3/Parametric Cooling/phonons_001x.npy")
ph_0005 = np.load("../../data/Chapter 3/Parametric Cooling/phonons_0005x.npy")

ax1.plot(t/(1/60e3), 293*ph_001/ph_001.max(), color=colors[0], alpha=0.5)
ax1.plot(t/(1/60e3), 293*ph_0005/ph_0005.max(), color=colors[0], linestyle='dashed')
ax1.set_xlim([0, t.max()/(1/60e3)])
ax1.grid(True, which="both", ls="-")
ax1.set_ylabel(r'$T_{\rm{eff}}^x$ (K)')
ax1.set_yscale('log')
ax1.set_yticks([1e-6, 1e-2, 1e2])

t = np.load("../../data/Chapter 3/Parametric Cooling/time_y.npy")
ph_001 = np.load("../../data/Chapter 3/Parametric Cooling/phonons_001y.npy")
ph_0005 = np.load("../../data/Chapter 3/Parametric Cooling/phonons_0005y.npy")
ax2.plot(t/(1/70e3), 293*ph_001/ph_001.max(), color=colors[0], alpha=0.5)
ax2.plot(t/(1/70e3), 293*ph_0005/ph_0005.max(), color=colors[0], linestyle='dashed')
ax2.set_xlim([0, t.max()/(1/70e3)])
ax2.grid(True, which="both", ls="-")
ax2.set_ylabel(r'$T_{\rm{eff}}^y$ (K)')
ax2.set_yscale('log')
ax2.set_yticks([1e-6, 1e-2, 1e2])


t = np.load("../../data/Chapter 3/Parametric Cooling/time_z.npy")
ph_001 = np.load("../../data/Chapter 3/Parametric Cooling/phonons_001z.npy")
t_shape = ph_001.shape[0]
ph_0005 = np.load("../../data/Chapter 3/Parametric Cooling/phonons_0005z.npy")
ax3.plot(np.linspace(0, 1000, t_shape), 293*ph_001/ph_001.max(), color=colors[0], alpha=0.5)
ax3.plot(np.linspace(0, 1000, t_shape), 293*ph_0005/ph_0005.max(), color=colors[0], linestyle='dashed')
ax3.set_xlim([0, 1000])
ax3.grid(True, which="both", ls="-")
ax3.set_ylabel(r'$T_{\rm{eff}}^z$ (K)')
ax3.set_yscale('log')#ax3.set_ylim([-1, 1])
ax3.set_yticks([1e-6, 1e-2, 1e2])
ax3.set_xlabel(r'$t^\ast$')
ax3.set_xticks([0, 250, 500, 750, 1000])
ax3.legend([r'$\Gamma_i=0.01\Omega_i$', r'$\Gamma_i=0.005Omega_i$'], loc='upper center', bbox_to_anchor=(0.5, 3.45),
          ncol=3, fancybox=True, shadow=True)

plt.subplots_adjust(wspace=0, hspace=0)
plt.savefig('../../plots/Chapter 3/parametric_control_pos.pdf', dpi=300, bbox_inches = 'tight')
plt.show()
```
