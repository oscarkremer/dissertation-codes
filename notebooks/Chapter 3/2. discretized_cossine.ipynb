```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 4, 100)
cossine_sig = np.cos((180*t)*np.pi/180)
```

```python
discretized_t, discretized_x = [], []
for i in range(cossine_sig.shape[0]):
    if not i%5:
        if len(discretized_x):
            discretized_t.append(t[i])
            discretized_x.append(discretized_x[-1])
        discretized_t.append(t[i])
        discretized_x.append(cossine_sig[i])
#    else:
#        discretized.append(discretized[-1])
```

```python
plt.plot(t, cossine_sig)
plt.plot(discretized_t, discretized_x)
```

```python
string_latex = ''
for t, u in zip(discretized_t, discretized_x):
    string_latex+= '({},{})--'.format(np.round(t+7.7,2), np.round(u,2))
```
